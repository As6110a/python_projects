```python
original_types = gdf['type_scene'].tolist()
original_types
```

```python
# append a new list for the data
new_types = []
print(new_types)
```

```python
# loop over entry to find index position of (::)
# then use slicing 

for s in original_types:
    # find start position of ('::')
    
    stop_idx = s.find('::')
    
    # get values up until stop index
    new_s = s[:stop_idx]
    
    # append new string to a list 
    new_types.append(new_s)
```

```python
# add a Boolean True/False column called accuracy

gdf['accuracy'] = gdf['answer']=='incorrect'
```

```python
gdf.groupby('type_scene_clean')['accuracy'].mean()
```

```python
## Omission errors--go trials--should have been pressed but wasn't: Incorrect

meangoRT = gdf.groupby('type_scene_clean')['accuracy'].mean()['basictrial'] * 100
meangoRT
```

```python
## Commission errors-nogo trials-shouldn't have been pressed but was: Incorrect still? 
## Or does now correct correspond to no go? Thought it's still incorrect since comission looks at false alarms?

gdf.groupby('type_scene_clean')['accuracy'].mean()['nogotrial'] * 100
```

```python
# use those values to get individual rates for SDT analysis

# percent correct on basictrials:
hits = gdf.groupby('type_scene_clean')['accuracy'].mean()['basictrial']

# misses is 1-hit_rate
misses = 1-gdf.groupby('type_scene_clean')['accuracy'].mean()['basictrial']

# percent correct on nogotrials (correct rejection)
correct_rej = gdf.groupby('type_scene_clean')['accuracy'].mean()['nogotrial']

# false alarm rate is 1-correct_rejections
false_alarms = 1-gdf.groupby('type_scene_clean')['accuracy'].mean()['nogotrial']


print(f'hit rate = {hits}; misses = {misses}; correct rejections = {correct_rej}; false alarms = {false_alarms}')
```

```python
## ssrt
## Reactive Inhibition: Calculated as SSRT mean across 25%, 50%
## Slower SSRTs indicate impaired reactive inhibition 
## Main outcome measure: (When p(respond | signal) equal to .50)  SSRT=mean go RT -
## observed mean SSD
```

```python
# calculate mean go RT in order to calculate the SSRT 
# get just stopsignalgotrials with a response (when the answer was correct + incorrect)

ssrt_trials = (gdf['type_scene_clean']=='stopsignalgotrial') & (gdf['answer'])
ssrt_df = gdf[ssrt_trials]
ssrt_df.head()
```

```python
# calculate SSRT as keypress time for correct stop signal go trials minus mean SSD 
 

ssrt = ssrt_df['key_press_time'].mean() - ssrt_df['thr_ssd_started'].mean()

ssrt

## Estimated that the stop process finished 417.9 ms? after the onset of the go stimulus. 
```

```python
## ssat: Stop-signal anticipation task 
## Measures both proactive (predictive cues indicate when to stop (goal-driven)) 
## and reactive inhibition (signal immediately triggers action cancellation (stimulus-driven))

## Stop-signal indicates to subject to withhold their response to an upcoming stimulus, 
## impaired proactive inhibition is apparent when there is a failure to increase RT when
## stop-signal probability increases (suggests a weaker anticipation of stopping)

## Reactive inhibition: calculated as SSRT mean across 25%, 50%
## Proactive inhibition: calculated as 0% GoRT minus (mean of 25% and 50% GoRT)

## Effect of stop-signal probability on response time (RT) (calculated
## by looking at covariance)
## Tendency to display slower response times as the probability of having to stop increases 
```

```python
ssat_trials = (gdf['type_scene_clean']=='anticipation50withsignal') & (gdf['answer'])
ssat_df = gdf[ssat_trials]
ssat_df.head()
```
